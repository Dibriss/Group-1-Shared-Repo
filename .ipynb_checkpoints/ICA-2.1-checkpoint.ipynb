## Part 2: Analyzing a 2D lattice model

The 2D lattice can be a nice visual simplication, but how can we apply it in a way that enables us to learn about physics? As with many physics phenomena, we can begin with energy.

The energy (also called Hamiltonian) of the entire lattice is

$$E_{\text{total}} = - \sum_{\langle a,b \rangle}  s_a s_b \text{,}$$

where $\langle a,b \rangle$ indicates all pairs of immediately adjacent spins $a$ and $b$. 

Most spins in the system will have **four adjacent neighbors**, except the spins on the outer edge of the lattice.

## Part 3: Monte Carlo modeling

We use computing to create models, in part because it provide capabilities like animation and randomization, which can't be achieved using pen and paper alone. One such model, that can only be created computationally (except for very simple examples), is a Monte Carlo simulation. Monte Carlo is a set of simulation algorithms that use probability distributions to model the evolution of a system.

We can use the [**Metropolis algorithm**](https://en.wikipedia.org/wiki/Metropolis%E2%80%93Hastings_algorithm) (a specific type of Monte Carlo method) to model a 2D lattice of spins.

The algorithm can be broken into these steps:

1. Establish an initial configuration of a system
2. Randomly generate a possible change to the system
3. Calculate the probability of accepting that change
4. Accept or reject the change based on that probability
5. Go back to Step 2 and repeat the algorithm

If the system being modeled is the 2D lattice of spins, we can model the probability like this:

$$\mathcal{P}(\text{accept}) = \begin{cases}
1 & \text{ if d}E \le 0 \\
e^{-\text{d}E/T} & \text{ if d}E > 0 
\end{cases}$$

This comes from the Boltzmann distribution.

&#x2705; Rewrite the five steps of the Metropolis algorithm for the 2D lattice model.

1. Establish initial conficuration of spins in a 2x2 lattice
2. choose a random orientation to flip
3. calculate change in energy of the system
4. Calculate probability of accepting or rejecting that change

5. Repeat 1-4
