## Part 2: Writing psuedocode

Once you have your implementation broken into steps, you can connect each of those pieces to the computing tools needed to build them. A great way to identify those computing tools and plan how to use them is pseudocode.

Here is a silly example to illustrate psuedocode. Suppose you are building a robot to chop onions for you because you don't want to cry while you are making dinner (and you forgot that goggles exist). 

You might plan to write some code for your robot using a structure like this:

You might write something like this:

```
Start with a bag of onions

For each onion in the bag:
    Put it on the cutting board
    Slice it up
    Put the slices into a saucepan
    
Cook up your onions
```

This is the idea behind pseudocode -- to map out your instructions for the computer using familiar coding structures (in this case, a `for` loop).

You can take this a step further by using coding terms a little more explicitly.

```
onion_bag = get_onion_bag()

for onion in onion_bag:
    place_food(onion, cutting_board)
    slices = slice_food(onion)
    place_food(slices, saucepan)
    
cook_food(saucepan, method="stir fry")
```

In the code above, we've replaced each line of psuedocode with actual code. Yes, take another look, this code could be executed in Python! There is a caveat here -- we used several undefined functions to make this translation:
- `get_onion_bag` is needed to initialize our bag of onions. We know that a bag of onions is needed, and we can declare this function later to tell our robot how to find the proper bag.
- `slice_food` is needed to tell our robot how to slice the onion. Note, we left this function pretty general, and we are inputting `onion` as an argument to the function. This leaves the door open to tell our robot to slice other foods in the future, for example maybe `slice_food(apple)`. We still need to declare this function later.
- `place_food` is needed for the mechanical movements of place onions onto a cutting board, and slices onto a saucepan. Because of the different objects and receptacles involved, we provide these as inputs to the function. This way, our function can be used for other similar tasks. Like the other functions used in this code, we'll need to declare this later.
- `cook_food` is the last function we use. For this function, we specified the cooking implement and the method. Our thinking was that there are many possible implements, so `saucepan` is provided as a regular input. However, there aren't a lot of cooking methods (at least, not that a robot could do), so we specified `method` as a default variable with only a few possible values, one of which could be `"stir fry"`.

Reading through the list above, you can tell we still have a lot of code to write:
- We need to declare four functions as described above.
- We also need to figure out how to define a couple variables we took for granted: `cutting_board`, `saucepan`.

However, this splits the remaining work up into six tasks, all of which seem significantly more simple and well-defined than our original task: to program a robot that can chop onions.

**This is the main power of pseudocode: to chunk up your work into much more manageable coding tasks.** This can be especially effective when you write it using placeholder functions like in the above example.
